# **Outlier Detection**

```python
selected=['Latitude','Longitude','price_mm','Reviews Per Month','Number Of Reviews']
```

```python
plt.figure(figsize=(12, 8))  # Set the figure size
num_rows = 2
num_cols = 3

for i, column in enumerate(selected, 1):
    plt.subplot(num_rows, num_cols, i)
    sns.boxplot(y=df[column],palette='rocket')
    plt.title(f'Box Plot of {column}')

plt.tight_layout()
plt.show()
```

```python
q1 = df['price_mm'].quantile(0.25)
q3 = df['price_mm'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
```

```python
q1,q3,iqr
```

```python
lower_bound,upper_bound
```

```python
outliers = df[(df['price_mm'] > upper_bound) | (df['price_mm'] < lower_bound)]
```

```python
outliers
```
